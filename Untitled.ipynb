```python
from research.strategies.macd import MACDStrategy, MACDParams
from common.data.yf_loader import load_daily
import pandas as pd

df = load_daily("HDFCBANK.NS", "2023-01-01", "2023-01-10")
print(df)
params = MACDParams(fast=6, slow=20, signal=5)
strat = MACDStrategy()

# Pick two days from df
days = list(df.index)
if len(days) >= 2:
    df_pair = df.loc[[days[0], days[1]]]
    print("Testing label for:", days[0], days[1])
    print(df_pair)
    print("PNL:", strat.run_day(df_pair, params))
```
